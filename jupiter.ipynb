```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches


# Chargement des donnÃ©es
data = pd.read_csv('Journalists.csv')  # Remplacez 'vos_donnees.csv' par le chemin vers vos donnÃ©es

#2. Effectif de l'Ã©chantillon
# Calculer l'effectif de l'Ã©chantillon
effectif_echantillon = data.shape[0]
print(f"Effectif de l'Ã©chantillon : {effectif_echantillon}")

#3. ReprÃ©sentation avec des stickmen de la population de journalistes tuÃ©s
#Nous allons utiliser les donnÃ©es pour crÃ©er une reprÃ©sentation simple avec des stickmen. Cependant, Matplotlib ne prend pas en charge directement les stickmen, mais nous pouvons utiliser des carrÃ©s ou des cercles pour reprÃ©senter les individus.
nb_petits_bonhommes = 1580

# CrÃ©ez une figure pour la visualisation
fig, ax = plt.subplots(figsize=(20, 30))

# Boucle pour dessiner chaque "petit bonhomme"
for i in range(nb_petits_bonhommes):
    # Calculez la position du "petit bonhomme" en fonction de sa position dans la liste
    x = (i % 40) * 0.5  # 40 "petits bonhommes" par ligne
    y = (i // 40) * 2.0  # Les lignes sont espacÃ©es de 2 unitÃ©s
    
    # Dessine la tÃªte
    plt.plot(x + 0.25, y + 1.5, marker='o', markersize=5, color='black')
    
    # Dessine le corps
    plt.plot([x + 0.25, x + 0.25], [y + 1.0, y + 0.5], color='black', lw=0.5)
    
    # Dessine les bras
    plt.plot([x + 0.25, x + 0.1], [y + 1.0, y + 0.8], color='black', lw=0.5)
    plt.plot([x + 0.25, x + 0.4], [y + 1.0, y + 0.8], color='black', lw=0.5)
    
    # Dessine les jambes
    plt.plot([x + 0.2, x + 0.2], [y + 0.5, y + 0.0], color='black', lw=0.5)
    plt.plot([x + 0.3, x + 0.3], [y + 0.5, y + 0.0], color='black', lw=0.5)

# DÃ©finissez les limites de l'axe pour que tous les "petits bonhommes" soient visibles
plt.xlim(0, 20)
plt.ylim(0, 60)

# Cachez les axes pour une visualisation plus clean
plt.axis('off')

# Affichez la figure
plt.show()
#4. % de la confirmation du motif
#Pour calculer le pourcentage de confirmation du motif, nous devons savoir combien de lignes ont un motif confirmÃ©. Nous supposerons ici que la colonne "Motive" est vide lorsque le motif n'est pas confirmÃ©.

# Calculer le pourcentage de confirmation du motif
motif_confirme = data['Motive'].notna().sum()
pourcentage_confirmation_motif = (motif_confirme / effectif_echantillon) * 100
print(f"% de confirmation du motif : {pourcentage_confirmation_motif:.2f}%")

#5. % hommes/femmes camembert 
# Calculer le pourcentage d'hommes et de femmes
hommes = data['Sex'] == 'Male'
femmes = data['Sex'] == 'Female'
pourcentage_hommes = (hommes.sum() / effectif_echantillon) * 100
pourcentage_femmes = (femmes.sum() / effectif_echantillon) * 100
print(f"% d'hommes : {pourcentage_hommes:.2f}%")
print(f"% de femmes : {pourcentage_femmes:.2f}%")


plt.pie([pourcentage_hommes, pourcentage_femmes], labels=['Hommes', 'Femmes'], autopct='%1.1f%%')
plt.title('RÃ©partition hommes/femmes/inconnu')
plt.axis('equal')  # Pour que le camembert soit circulaire
plt.show()

#6. RÃ©partition des morts par annÃ©e
# Calculer la rÃ©partition des morts par annÃ©e
morts_par_annee = data['Year of Death'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
morts_par_annee.plot(kind='bar')
plt.title("RÃ©partition des morts par annÃ©e")
plt.xlabel("AnnÃ©e")
plt.ylabel("Nombre de morts")
plt.show()
```
